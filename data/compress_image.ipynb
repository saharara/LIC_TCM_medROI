```python
import gzip
import shutil
import os

path = './data/nifti'
for filename in os.listdir(path):
    if filename.endswith('.nii'):
        with open(os.path.join(path, filename), 'rb') as f_in:
            with gzip.open(os.path.join(path, filename + '.gz'), 'wb') as f_out:
                shutil.copyfileobj(f_in, f_out)
        os.remove(os.path.join(path, filename)) # Xóa file cũ sau khi nén
        print(f"Đã nén: {filename} -> {filename}.gz")
```
